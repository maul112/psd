## Install KNIME dan Python Script

Kali ini saya akan memberikan cara menggunakan software KNIME sederhana. yang pertama kita harus download software KNIME di url https://www.knime.com/downloads lalu install. Tampilan software KNIME awal-awal akan seperti berikut:

![Teks alternatif](img/Screenshot%20(236).png)

Kali ini saya ingin menggunakan Python Script. Awal-awal kalian tidak akan menemukan Python Script pada KNIME dibagian kategori Scripting, adanya cuman Java. Maka dari itu kita install terlebih dahulu Python Script nya caranya adalah:

1. Klik menu di pojok kanan atas, lalu pilih Install Extention.

![Teks alternatif](img/Screenshot%20(237).png)

2. Lalu cari Python (disini tidak ada karena saya sudah install).

![Teks alternatif](img/Screenshot%20(239).png)

pilih yang namanya sesuai sama berikut:

![Teks alternatif](img/Screenshot%202025-10-08%20111926.png)

Jika sudah install, KNIME akan meminta untuk restart software. Setelah itu jika sudah muncul node Python berarti proses instalasi berhasil.

## Menarik data dari PostgreSQL ke KNIME

1. Drag node "PostgreSQL Connector" kedalam Workflow.

![Teks alternatif](img/Screenshot%20(241).png)

2. Lalu klik 2 kali pada node tersebut untuk konfigurasi. Sesuaikan data-data dengan konfigurasi PostgreSQL yang kalian miliki.

![Teks alternatif](img/Screenshot%20(242).png)

3. Lalu kita butuh konfigurasi driver PosstgreSQL di KNIME. Klik Preference pada halaman utama KNIME. Lalu pilih databases lalu klik add.

![Teks alternatif](img/Screenshot%20(248).png)

Pilih file driver PostgreSQL dengan cara klik add file. Setelah menambahkan driver klik Find Driver Classes lalu klik Ok (download driver di url https://jdbc.postgresql.org/download/).

![Teks alternatif](img/Screenshot%20(249).png)

4. Ketika ingin memilih tabel, gunakan node "DB Table Selector" lalu tuliskan schema dan nama tabel.

![Teks alternatif](img/Screenshot%20(243).png)

![Teks alternatif](img/Screenshot%20(244).png)

5. Setelah itu, sambungkan antar node seperti digambar (drag kotak pink kecil samping node PostgreSQL Connector dan sambungkan ke kotak pink di node DB Table Selector)

![Teks alternatif](img/Screenshot%20(245).png)

6. Untuk membaca database kita perlu menambahkan node "DB Reader".

![Teks alternatif](img/Screenshot%20(246).png)

7. Lalu run semua node hingga menunjukkan lingkaran hijau kecil dibawah node.

![Teks alternatif](img/Screenshot%20(250).png)

8. Lalu kita tambahkan node Python Script (Legacy).

![Teks alternatif](img/Screenshot%20(251).png)

9. Lalu konfigurasi Python dengan cara pergi ke Preferences → KNIME → Python (Konfigurasi juga di Python (Legacy)). Pilih manual lalu pilih path python yang anda gunakan.

![Teks alternatif](img/Screenshot%20(252).png)

![Teks alternatif](img/Screenshot%20(253).png)

10. Lalu run node Python. Pada gambar terlihat sudah keluar untuk tampilan tabel yang sudah kita tarik dari PostgreSQL ke KNIME

![Teks alternatif](img/Screenshot%20(254).png)
